```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

dir_path = os.getcwd()
plt.style.use("bmh")
```

## Load Full ETFs Dataset

```python
etfs_df = pd.read_csv(os.path.join(dir_path, "data", "inputs", "etfs.csv"))
etfs_df["date"] = pd.to_datetime(etfs_df["date"])
etfs_df.set_index("date", inplace=True)

etfs_df.head()
```

## Build Daily Data

```python
daily_etfs_df = etfs_df.resample("B").ffill()

daily_etfs_df.head()
```

```python
daily_etfs_df.to_csv(os.path.join(dir_path, "data", "inputs", "daily_etfs.csv"), index=True)
```

## Build Daily Returns Data

```python
rets_etfs_df = np.log(etfs_df).diff()

rets_etfs_df.head()
```

```python
rets_etfs_df.to_csv(os.path.join(dir_path, "data", "inputs", "daily_ret_etfs.csv"), index=True)
```

## Build Monthly Data

```python
monthly_etfs_df = etfs_df.resample("B").ffill().resample("BMS").first()

monthly_etfs_df.tail()
```

```python
monthly_etfs_df.to_csv(os.path.join(dir_path, "data", "inputs", "monthly_etfs.csv"), index=True)
```

## Build Monthly Returns Data

```python
monthly_ret_etfs_df = np.log(etfs_df.resample("B").ffill()).diff(25).resample("BMS").first()

monthly_ret_etfs_df.tail()
```

```python
monthly_ret_etfs_df.to_csv(os.path.join(dir_path, "data", "inputs", "monthly_ret_etfs.csv"), index=True)
```
